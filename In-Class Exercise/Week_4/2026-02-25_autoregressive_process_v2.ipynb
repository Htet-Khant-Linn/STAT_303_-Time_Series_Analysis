```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

Creating White Noise

```python
np.random.seed(0)
ts_steps = 300
ts_value = np.random.normal(loc=0, scale=1, size=ts_steps)
ts = pd.date_range(start='2026-01-01', periods=ts_steps, freq='D')

df_white_noise = pd.DataFrame(data={'data':ts_value}, index=ts)
```

```python
df_white_noise.info()
```

```python
df_white_noise.head()
```

```python
ax = df_white_noise.plot(figsize=[12, 5])
ax.set_title('White Noise')
ax.set_xlabel('Time')
ax.set_ylabel('Data')
```

```python
fig, ax = plt.subplots(figsize=[12, 5])
plot_acf(
    x=df_white_noise['data'],
    lags=50,
    ax=ax,
    alpha=0.05,
    auto_ylims=True
);
ax.set_title('Autocorrelation of white noise')
ax.set_ylabel('Autocorrelation')
ax.set_xlabel('Lag')
```

```python
fig, ax = plt.subplots(figsize=[12, 5])
plot_pacf(
    x=df_white_noise['data'],
    method='ywmle',
    lags=50,
    ax=ax,
    alpha=0.05,
    auto_ylims=True
);

ax.set_title('Partial Autocorrelation of white noise')
ax.set_ylabel('Partial Autocorrelation')
ax.set_xlabel('Lag')
plt.tight_layout()
```

## Generate AR(1) Process

```python
# Generate AR(1) process
num_timesteps = 300
phi_1 = 0.9 # larger the phi value, larger the correlation value
ts = pd.date_range(start="2026-01-01", periods=num_timesteps, freq="D")

y = np.zeros(num_timesteps)

for t in range(1, num_timesteps):
    noise = np.random.normal()
    y[t] = phi_1 * y[t-1] + noise

df_ar1 = pd.DataFrame(data={'y':y}, index=ts)
```

```python
df_ar1.head()
```

## Create AR(1) Plot

```python
ax = df_ar1.plot(figsize=[12,5])
ax.set_title("AR (1) Process")
ax.set_ylabel("y")
ax.set_xlabel("Time")
```

### ACF & PACF

```python
fig, ax = plt.subplots(figsize=[10,5])
plot_acf(
    x = df_ar1['y'],
    lags=36,
    ax=ax,
    alpha=0.05,
    auto_ylims=True
);
ax.set_title("Parial autocorrelation of AR(1) process")
ax.set_ylabel("Partial autcorrelation")
ax.set_xlabel("Lag")
plt.tight_layout()

fig,ax = plt.subplots(figsize=[10,5])
plot_pacf(
    x=df_ar1,
    method= "ywmle",
    lags = 50,
    ax = ax,
    alpha=0.05,
    auto_ylims=True
    );
ax.set_title("Parial autocorrelation of AR(1) process")
ax.set_ylabel("Partial autcorrelation")
ax.set_xlabel("Lag")
plt.tight_layout()
```

## Generate AR(2) process

```python
phi_1 = 0.7
phi_2 = 0.3
y = np.zeros(num_timesteps)

for t in range(1, num_timesteps):
    noise = np.random.normal()
    y[t] = phi_1 * y[t - 1] + phi_2 * y[t - 2] + noise

df_ar2 = pd.DataFrame(data={"y":y}, index=ts)
```

```python
df_ar2.head()
```

# Plot AR(2) process

```python
ax = df_ar2.plot(figsize=[10, 5])
ax.set_title("AR(2) process")
ax.set_ylabel("y")
ax.set_xlabel("Time")

fig, ax = plt.subplots(figsize=[10,5])
plot_acf(
    x = df_ar2,
    lags= 36,
    ax = ax,
    alpha=0.05,
    auto_ylims=True
    );
ax.set_title("Autocorrelation of AR(2) process")
ax.set_ylabel("Autocorrelation")
ax.set_xlabel("Lag")

fig,ax = plt.subplots(figsize=[10,5])
plot_pacf(
    x=df_ar2,
    method= "ywmle",
    lags = 50,
    ax = ax,
    alpha=0.05,
    auto_ylims=True
    );
ax.set_title("Parial autocorrelation of AR(2) process")
ax.set_ylabel("Partial autcorrelation")
ax.set_xlabel("Lag")
plt.tight_layout()
```
