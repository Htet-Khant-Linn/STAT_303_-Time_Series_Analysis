```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
num_timesteps = 300
np.random.seed(0)

data = np.random.normal(loc=0, scale=1, size=num_timesteps)

ts = pd.date_range(start='2026-01-01', periods=num_timesteps, freq="D")
```

```python
df = pd.DataFrame(data={'data':data}, index=ts)
print(df.info())
print(df.head())
```

```python
retial_sales = pd.read_csv('data/retail_sales.csv', parse_dates=['ds'], index_col='ds')
print(retial_sales.info())
```

```python
plt.figure(figsize=(12, 6))
retial_sales.y.plot(color = 'blue', linewidth=0.8)
plt.title('Time Series Plot of Retail Sales')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend()

plt.tight_layout()

plt.show()
```

```python
# plt.figure(figsize=(12, 6))
plot_acf(
    x = retial_sales.y,
    lags=12,
    alpha=0.05,
    auto_ylims=True
);
plt.title('Auto Correlation Plot of Retail Sales')
plt.xlabel('Time')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```

```python
# plt.figure(figsize=(12, 6))
plot_pacf(
    x = retial_sales.y,
    method='ywmle',
    lags=12,
    alpha=0.05,
    auto_ylims=True
);
plt.title('Partial Auto-Correlation Plot of Retail Sales')
plt.xlabel('Time')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```
