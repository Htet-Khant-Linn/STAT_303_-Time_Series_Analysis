```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv('data\global_temp_index.csv')
```

```python
Data.info()
```

```python
Data.head()
```

```python
year = Data.iloc[:, 0]
year
```

```python
temp_index = Data.iloc[:, 1]
temp_index
```

```python
plt.plot(year, temp_index, "r")
plt.xlabel('Year', family = 'Times New Roman')
plt.xticks(year, rotation = 45) # adding data points and rotating 45 degree
plt.ylabel('Temperature Index', family = 'Times New Roman')
plt.title('Global Warming', family = 'Times New Roman')
plt.tight_layout()

plt.savefig('outputs\Global_Warming.png') # we need to save before show()

plt.show()
```
