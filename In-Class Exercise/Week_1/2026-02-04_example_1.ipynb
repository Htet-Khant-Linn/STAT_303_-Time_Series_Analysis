## Multiplication

```python
first_array = np.array([[1,2,3],[4,5,6]])
print(first_array)
print()
A = np.array([[1,2], [3,4]])
B = np.array([[3, 4], [5,7]])
print(A)
print()
print(B)
print()

C = A @ B   # matric multiplication
print(C)
print()

C_element = A * B
print(C_element)
print()

D = np.dot(A, B)    # matrix multiplication
print(D)
print()

E = np.multiply(A, B)
print(E)
print()
```

## Adding and substraction

```python
sum1 = A + B
print("A + B")
print(sum1)
print()

sub1 = A - B
print("A - B")
print(sub1)
print()

sub2 = np.subtract(A, B)
print("np.subtract(A, B)")
print(sub2)
print()

sum2 = np.sum(A)
print("np.sum(A)")
print(sum2)
print()

Borad_num = A + 3
print("A + 3")
print(Borad_num)
print()

board_matrix = np.array([[3,3], [3,3]])
```

## Dividing

```python
D = np.divide([12, 14,15], 5)
print("np.divide([12, 14,15], 5)")
print(D)
print()

D_floor = np.floor_divide([12, 14, 15], 5)
print("np.floor_divide([12, 14, 15], 5)")
print(D_floor)
print()

sq = np.sqrt(10)
print("np.sqrt(10)")
print(sq)
print()
```
