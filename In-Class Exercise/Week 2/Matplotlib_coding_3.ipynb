```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv("data\montly_electricity_consumption.csv")
```

```python
Data.info()
```

```python
Data.head()
```

```python
month = Data.loc[:, 'month']
month
```

```python
customerA = Data.loc[:, 'customer_A_kwh']
customerA
```

```python
customerB = Data.loc[:, 'customer_B_kwh']
customerB
```

## Creating Line Plot

```python
plt.plot(month, customerA)
plt.plot(month, customerB)
plt.xticks(month, rotation = 45)
plt.tight_layout()
plt.show()
```

```python
plt.plot(month, customerA, color = 'red', label = 'Customer A', marker = 'o')
plt.plot(month, customerB, color = 'blue', label = 'Customer B', marker = '*')
plt.xlabel('Month')
plt.ylabel('kWh')
plt.title('Electricity Consumption Chart')
plt.xticks(month, rotation = 45)
plt.legend()
plt.tight_layout()

plt.savefig('outputs\Electricity_Consumption_Chart.png')

plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(month, customerA, color = 'red', label = 'Customer A', marker = 'o')
plt.xlabel('Month')
plt.ylabel('kWh')
plt.xticks(month, rotation = 45)

plt.subplot(2, 1, 2)
plt.plot(month, customerB, color = 'blue', label = 'Customer B', marker = '*')
plt.xlabel('Month')
plt.ylabel('kWh')
plt.xticks(month, rotation = 45)
plt.suptitle('Electricity Consumption Chart')

plt.tight_layout()
plt.legend()

plt.savefig('outputs\Electricity_Consumption_Chart_sub_plot.png')

plt.show()
```

## Creating Scatter Plot

```python
plt.scatter(month, customerA, color = 'red', label = 'Customer A', marker = 'o')
plt.scatter(month, customerB, color = 'blue', label = 'Customer B', marker = '*')
plt.xlabel('Month')
plt.ylabel('kWh')
plt.title('Electricity Consumption Chart')
plt.legend(loc = 'best')
# plt.legend(loc = (0.1, 0.1))

plt.grid()
plt.tight_layout()

plt.savefig('outputs\Electricity_Consumption_Chart_scatter_plot.png')

plt.show()
```

```python
plt.hist(customerA, bins = 20, color = 'blue', rwidth = 0.8)
plt.xlabel('kWh')
plt.ylabel('frequency')
plt.title('Electricity Consumption Histogram')



plt.show()
```

```python
plt.bar(month, customerA, color = 'blue')
plt.xlabel('Month')
plt.ylabel('kWh')
plt.title('Electricity Consumption Bar Graph')
plt.show()
```
