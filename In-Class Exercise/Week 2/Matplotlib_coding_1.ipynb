# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
print(x)

'''
Key Parameters
The primary parameters for np.linspace() are: 
start: The beginning value of the sequence.
stop: The end value of the sequence. By default, this value is included in the array.
num: (Optional) The number of evenly spaced samples to generate. The default is 50. 
'''
```

```python
y = x ** 2
print(y)
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Lavel')
plt.title('Title')
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Scattre Plot')
plt.show()
```

## Creating Subplots

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title 1')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title 2')

plt.suptitle('Main Title')

plt.savefig('outputs\my_figure.png', dpi = 300)

plt.figure() # this seperate the two visual otherwise it will overlap
plt.plot(x, x**2, label = 'x_squared')
plt.plot(x, x**3, label = 'x cubed')
plt.title('Title')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc = 'best')
plt.show()
```
