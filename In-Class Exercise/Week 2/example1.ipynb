```python
import pandas as pd
```

# Series

```python
Age = pd.Series([10, 20, 30, 40], index = ['age1', 'age2', 'age3', 'age4'])
```

```python
print(Age.age3)
```

```python
print(Age.age1)
```

```python
# Filtering values of the Series
Filtered_Age = Age[Age>10]
print(Filtered_Age)
```

```python
Filtered_Age_2 = Age[(Age > 10) & (Age < 40)]
print(Filtered_Age_2)
```

```python
Filtered_Age_3 = Age[Age.between(20, 30)]
print(Filtered_Age_3)
```

```python
# Calling values of the Series
Age.values
```

```python
# Calling indexes of the Series
Age.index
```

# DataFrame

```python
import numpy as np
```

```python
# Creating DataFrame

DF = np.array([[20, 10, 8], [25, 8, 10], [27, 5, 3], [30, 9, 7]])
# Creating array first

print(DF)
type(DF)
```

```python
Data_set = pd.DataFrame(DF)
print(Data_set)
type(Data_set)
```

```python
Data_set = pd.DataFrame(DF, index = ['S1', 'S2', 'S3', 'S4'])

print(Data_set)
```

```python
Data_set = pd.DataFrame(DF, index = ['S1', 'S2', 'S3', 'S4'], columns = ['Age', 'Grade1', 'Grade2'])

print(Data_set)
```

```python
# Adding another column

Data_set['Grade3'] = [9, 6, 7, 10]
print(Data_set)
```

# Indexing the DataFrame

**.loc and iloc**

.loc is label-based indexing and include the end

.iloc is integer-based indexing and exclude the end

```python
Data_set.loc['S1']

# (Single Brackets): Returns a pandas Series. This is a one-dimensional array representing just that single row of data.
```

```python
Data_set.loc[['S1']]

# (Double Brackets): Returns a pandas DataFrame. By passing a list ['S1'] inside the brackets, pandas keeps the table structure, resulting in a 2-dimensional table that happens to have only one row.
```

```python
Data_set.loc['S1':'S2']
```

```python
Data_set.iloc[1, 2]
```

```python
Data_set.iloc[:, 0]
# all row and first column
```

```python
Data_set.iloc[:, 3]
```

```python
Data_set.iloc[:, 1:3]   # doesn't not include the end
```

```python
Data_set.iloc[:, :3]
```

```python
Data_set.iloc[1:2, :]
```

```python
Data_set.iloc[:3, :]
```

```python
Data_set.iloc[:, :-1]
```
